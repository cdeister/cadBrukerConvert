```python
pixelsPerLine=512
lineCount=512
multSamp=3
pixelsPerFrame=int(pixelsPerLine*lineCount*multSamp)
framesInChunk=rd.size/pixelsPerFrame
print(framesInChunk)
```

```python
# The frames from a resonant scan are 2x pixels per line, but each pixel can be sampled more than once.
# Each sequential line is a flipped version of what it should be. 
# This scan was 3x per pixel.

pixelsPerLine=512
pixelsPerLineScan=pixelsPerLine*2
multisamplePixels=3

curPixels=rd[0:pixelsPerLineScan*multisamplePixels*int(lineCount/2)]
curFrame=np.reshape(curPixels,(int(lineCount/2),pixelsPerLineScan*multisamplePixels))
curFrame.shape
```

```python
cfPlot=plt.imshow(curFrame,aspect='auto')
```

```python
# Now we need to split and flip every other.
# a) Split
curFrame=np.reshape(curFrame,(512,1536))
cfPlot=plt.imshow(curFrame,aspect='auto')
```

```python
# b) flip every other
curFrame[0:512:2,:]=np.fliplr(curFrame[0:512:2,:])
cfPlot=plt.imshow(curFrame,aspect='auto')
```

```python
def convertResFrame(rawFrame,linePx,tLines,sampPerPx):
    
    fullResLinePx=int(linePx*sampPerPx*2)
    resLines=int(tLines*0.5)
    halfResLinePx=int(linePx*sampPerPx)
    cFr=np.reshape(rawFrame,(resLines,fullResLinePx))
    cFr=np.reshape(cFr,(tLines,int(linePx*sampPerPx)))
    cFr[0:halfResLinePx:2,:]=np.fliplr(cFr[0:halfResLinePx:2,:])

    return cFr
```

```python
# Lets combine it all for any frame.
dset=np.zeros((lineCount,int(lineCount*multisamplePixels),int(framesInChunk)))
tPx=pixelsPerLineScan*multisamplePixels*int(lineCount/2)
for ci in range(0,int(framesInChunk)):
    cPX=rd[0+(ci*tPx):tPx+(ci*tPx)]
    dset[:,:,ci]=convertResFrame(cPX,512,512,3)
    if np.mod(ci,200)==0:
        print("fin: {}".format(ci))
```

```python
dSetMean=dset[:,0:1536:3,:]+dset[:,1:1536:3,:]+dset[:,2:1536:3,:]
```
